Notre corpus présenté au format `csv` peut aussi être ouvert avec une bibliothèque python

On teste d'ouvrir notre corpus au format csv avec la bibliothèque `csv`.

```python
import csv

with open("../data/corpus.csv", 'r') as file:
    data = csv.reader(file, delimiter=";")
    for row in data:
        print(row)
```

La bibliothèque `datasets` permet de charger facilement les données des corpus disponibles sur le hub de `hugging face` mais est aussi compatible avec la bibliothèque `pandas`

On peut donc tester d'ouvrir notre corpus PLOD/CW avec datasets

```python
from datasets import Dataset

dataset = load_dataset("surrey-nlp/PLOD-CW", split="train")
dataset = dataset.to_pandas()
dataset = Dataset.from_pandas(dataset)
```

On ouvre les informations sur le dataset PLOD/CW

```python
dataset
```
